```python
hasilgabung = pd.concat([pq,pr,ps],ignore_index=True)
```

```python
hasilgabung.columns.values
```

```python
x = hasilgabung.iloc[:,7: 76]
```

```python
x
```

```python
y = hasilgabung.iloc[:,83: 84]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 43)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
asha = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
asha.fit(x_train,y_train)
```

```python
y_predic = asha.predict(x_test)
```

```python
y_predic
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_predic)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 8))
tree.plot_tree(asha, feature_names = x.columns.values, class_names = np.array([ 'pq','pr','ps']), filled = True)
plt.show()
```

```python
import seaborn as yey
from sklearn import metrics
label = np.array([ 'pq','pr','ps'])
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(y_test, y_predic)
plt.figure(figsize=(14, 14))
yey.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
